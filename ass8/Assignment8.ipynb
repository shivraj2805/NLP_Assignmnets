```python
import nltk
nltk.download('wordnet')
nltk.download('punkt')
```

```python
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize

sentence = "He went to the bank to deposit money"
word = "bank"

sense = lesk(word_tokenize(sentence), word)

print("Sentence:", sentence)
print("Best Sense:", sense)
print("Meaning:", sense.definition())
```

```python
from nltk.corpus import wordnet as wn
```

```python
sentence = "The fisherman sat on the bank of the river"
word = "bank"

context = set(word_tokenize(sentence.lower()))

best_sense = None
max_overlap = 0

for sense in wn.synsets(word, pos=wn.NOUN):
    definition = sense.definition()
    def_tokens = set(word_tokenize(definition.lower()))
    
    overlap = len(context.intersection(def_tokens))
    
    if overlap > max_overlap:
        max_overlap = overlap
        best_sense = sense

print("Sentence:", sentence)
print("Best Sense:", best_sense)
print("Meaning:", best_sense.definition())
```
