```python
import nltk

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
text = "NLTK is amazing! I can't believe it's working so well. #AI @OpenAI loves NLP :)"
```

```python
# WHite space tokenization
from nltk.tokenize import WhitespaceTokenizer

wt = WhitespaceTokenizer()
print(wt.tokenize(text))
```

```python
# Punctuation-based Tokenization
from nltk.tokenize import wordpunct_tokenize

print(wordpunct_tokenize(text))
```

```python
#Treebank Tokenization
from nltk.tokenize import TreebankWordTokenizer

tbt = TreebankWordTokenizer()
print(tbt.tokenize(text))
```

```python
#Tweet Tokenization
from nltk.tokenize import TweetTokenizer

tt = TweetTokenizer()
print(tt.tokenize(text))
```

```python
# MWE (Multi-Word Expression) Tokenization
from nltk.tokenize import MWETokenizer

mwe = MWETokenizer([('New', 'York'), ('Artificial', 'Intelligence')], separator='_')
sentence = "I love Artificial Intelligence and New York city"
print(mwe.tokenize(sentence.split()))
```

Stemming

```python
#Porter Stemmer
from nltk.stem import PorterStemmer

ps = PorterStemmer()
words = ["running", "runs", "easily", "fairly"]

print([ps.stem(word) for word in words])
```

```python
#Snowball Stemmer
from nltk.stem import SnowballStemmer

ss = SnowballStemmer("english")
print([ss.stem(word) for word in words])
```

Lemmatization (WordNet Lemmatizer)

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("running", pos="v"))
print(lemmatizer.lemmatize("better", pos="a"))
print(lemmatizer.lemmatize("cars", pos="n"))
```
